```python
df = pd.read_csv("이미지_색감_비율.csv")
df
```

```python
data = pd.read_csv("https붙인_img.csv")
data
```

```python
print(len(data['img'].unique()))
print(len(df['img'].unique()))
```

```python
df_unique = df.iloc[::3]
df_unique = df_unique.reset_index(drop=True)
df_unique
```

```python
img_list = []
object_list = []
for i in range(len(df_unique)):
    img_list.append(df_unique['img'][i])
    object_list.append(df_unique['이미지_객체'][i])
```

```python
print(len(img_list))
print(len(object_list))
```

```python
data['curObject'] = data['img']
data.head(5)
```

```python
img = img_list
curObject = object_list
obj_dic = dict(zip(img,curObject))
obj_dic
```

```python
data['curObject'] = data['curObject'].map(obj_dic)
data
```

```python
data.to_csv('객체정보 수정.csv',index=False,encoding='utf-8-sig')
```
