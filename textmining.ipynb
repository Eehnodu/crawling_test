```python
import pandas as pd
import csv
# 토큰 갯수세기
from collections import Counter
import re
# 반복 작업을 시각화 시켜주는 도구
from tqdm import tqdm

from konlpy.tag import Okt
```

```python
data = pd.read_csv('curation_data/merge_data.csv')
data
```

```python
# 각 범위 내 첫 번째 행만 남기고 삭제할 인덱스 계산
indexes_to_drop = []

# 예를 들어 6개의 행이 있고, 각 범위가 3행이라고 할 때,
# 0~2에서 1, 2 삭제하고, 3~5에서 4, 5를 삭제
for i in range(0, len(data), 3):  # 여기서 3은 범위의 크기입니다.
    if i + 1 < len(data): indexes_to_drop.append(i + 1)
    if i + 2 < len(data): indexes_to_drop.append(i + 2)

# 계산된 인덱스에 해당하는 행 삭제
data = data.drop(indexes_to_drop)
data
```

```python
title_text = data['제목']
explain_text = data['설명']
```

```python
my_okt = Okt()  # 형태소분석기 객체 생성
```

```python
title_morphs = []
for txt in tqdm(title_text):  # 훈련용 문장을 반복
    morphs = my_okt.pos(str(txt))  # 각 문장을 형태소와 품사 태그로 분리
    filtered_morphs = [word for word, tag in morphs if tag in ['Noun', 'Adjective']]  # 명사와 형용사만 필터링
    title_morphs.append(filtered_morphs)
```

```python
title_morphs
```

```python
explain_morphs = []
for txt in tqdm(explain_text):  # 훈련용 문장을 반복
    if pd.isna(txt):  # NaN 값을 건너뛰기
        continue
    morphs = my_okt.pos(txt)  # 각 문장을 형태소와 품사 태그로 분리
    filtered_morphs = [word for word, tag in morphs if tag in ['Noun', 'Adjective']]  # 명사와 형용사만 필터링
    explain_morphs.append(filtered_morphs)
```

```python
explain_morphs
```

```python
# 한글자 제거 및 정규표현식 활용 불필요한 토큰 삭제
import re # 정규표현식
p = re.compile("[ㅋㅇㅎㄷ~ㅡ>:#!.?,0-9a-z]+")  # 정규표현식 패턴
```

```python
title_clean_text = []  # 전처리된 결과가 들어갈 리스트
for sentence in tqdm(title_morphs):  # 한 문장씩 꺼내오는 반복문
    tmp = []
    for token in sentence :   # 형태소 단위로 쪼개진 토큰을 반복

        if p.match(token):  # 정규표현식에 매칭되는 글자 제거
            continue
        tmp.append(token)  # 나머지 정상 토큰을 리스트에 등록
    title_clean_text.append(tmp)
```

```python
explain_clean_text = []
for sentence in tqdm(explain_morphs):  # 한 문장씩 꺼내오는 반복문
    tmp = []
    for token in sentence :   # 형태소 단위로 쪼개진 토큰을 반복
        if p.match(token):  # 정규표현식에 매칭되는 글자 제거
            continue
        tmp.append(token)  # 나머지 정상 토큰을 리스트에 등록
    explain_clean_text.append(tmp)
```

```python
# 하나의 리스트로 모아주기
title_token_list = []
explain_token_list = []
for s in title_clean_text:
    title_token_list += s
for s in explain_clean_text:
    explain_token_list += s
```

```python
title_counter = Counter(title_token_list)
```

```python
title_counter
```

```python
explain_counter = Counter(explain_token_list)
```

```python
explain_counter
```

```python
# Counter 객체를 DataFrame으로 변환
title_df = pd.DataFrame(title_counter.items(), columns=['단어', '빈도'])
# 빈도수 순으로 정렬
title_df = title_df.sort_values(by='빈도', ascending=False).reset_index(drop=True)
# 결과 확인
title_df  # 처음 몇 개의 행을 출력하여 확인
```

```python
# Counter 객체를 DataFrame으로 변환
explain_df = pd.DataFrame(explain_counter.items(), columns=['단어', '빈도'])
# 빈도수 순으로 정렬
explain_df = explain_df.sort_values(by='빈도', ascending=False).reset_index(drop=True)
# 결과 확인
explain_df  # 처음 몇 개의 행을 출력하여 확인
```

```python
title_df.to_csv("curation_data/제목 단어 빈도수.csv", index=False, encoding="utf-8-sig")
explain_df.to_csv("curation_data/설명 빈도수.csv", index=False, encoding="utf-8-sig")
```
