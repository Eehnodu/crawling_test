```python
## 실제로 검색을 해보니 가수명, 곡명을 검색을 하면 일치하는 정보가 하나만 뜸
```

```python
driver = wb.Chrome()
driver.get('https://www.melon.com/search/total/index.htm?q=vos+%EB%B6%80%EB%94%94&section=&searchGnbYn=Y&kkoSpl=Y&kkoDpType=&mwkLogType=T')

lyrics_list = []
artist_list = []
album_list = []
release_list = []
genre_list = []

for index, song in enumerate(tqdm(song_list)):
    # 검색어 입력
    search = driver.find_element(By.CSS_SELECTOR, "#top_search")
    search.clear()
    search.send_keys(f'{song}')
    
    if index == 0:
        # 첫 번째 검색에 사용할 버튼
        btn = wait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#header_wrap > div.wrap_search_field > fieldset > button.btn_icon.search_m')))
    else:
        # 두 번째 검색부터 사용할 버튼
        btn = wait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#gnb > fieldset > button.btn_icon.search_m')))
    
    btn.click()
    time.sleep(1.5)

    try :
        # 곡 정보 페이지 이동
        btn = driver.find_element(By.CSS_SELECTOR, '#frm_songList > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > div > div > a.btn.btn_icon_detail')
        btn.click()

        # 가사
        lyrics = driver.find_element(By.CSS_SELECTOR, '#lyricArea').text
        if (lyrics.startswith('[가사 준비중]')):
            lyrics = ''

        # 곡명
        song_name = driver.find_element(By.CSS_SELECTOR, "#downloadfrm > div > div > div.entry > div.info > div.song_name").text
        # 아티스트명
        artist = driver.find_element(By.CSS_SELECTOR, "#downloadfrm > div > div > div.entry > div.info > div.artist").text
        # 앨범
        album = driver.find_element(By.CSS_SELECTOR, "#downloadfrm > div > div > div.entry > div.meta > dl > dd:nth-child(2)").text
        # 발매일
        release = driver.find_element(By.CSS_SELECTOR, "#downloadfrm > div > div > div.entry > div.meta > dl > dd:nth-child(4)").text
        # 장르
        genre = driver.find_element(By.CSS_SELECTOR, "#downloadfrm > div > div > div.entry > div.meta > dl > dd:nth-child(6)").text
        
        lyrics_list.append(lyrics)
        artist_list.append(artist)
        album_list.append(album)
        release_list.append(release)
        genre_list.append(genre)
        
    except :
        continue
    time.sleep(2)
```
