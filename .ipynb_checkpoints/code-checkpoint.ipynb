```python
from tqdm import tqdm
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
import time
import re  # 정규 표현식 모듈 추가

# webdriver options
options = wb.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('disable-dev-shm-usage')

# 이미지 로딩 비활성화
options.add_argument('--disable-images')
options.add_experimental_option("prefs", {'profile.managed_default_content_settings.images': 2})
options.add_argument('--blink-settings=imagesEnabled=false')

# YouTube 채널 페이지 요청 및 동영상 링크 수집 코드
driver = wb.Chrome(options=options)

driver.get("https://www.youtube.com/@essentialme/videos")
time.sleep(2)

body = driver.find_element(By.TAG_NAME, "body")
last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    body.send_keys(Keys.END)
    time.sleep(2)
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

thumbnails = driver.find_elements(By.CSS_SELECTOR, "a#thumbnail")

link_set = set()
for thumb in thumbnails:
    href = thumb.get_attribute('href')
    if href is not None:
        link_set.add(href)

link_list = list(link_set)

# 모든 링크를 저장할 리스트
all_links = []

# 이미지를 저장할 리스트
img_list = []

# 영상의 제목을 저장할 리스트
head_list = []

# 각 동영상 페이지에서 설명란 링크 추출
for video_url in tqdm(link_list):
    driver.get(video_url)
    time.sleep(3)

    # '더보기' 버튼 클릭
    try:
        thebogi = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "#expand"))
        )
        thebogi.click()
        time.sleep(2)
    except NoSuchElementException as e:
        print(f"더보기 버튼을 찾을 수 없습니다: {e}")
    
    # 제목에 [Playlist]가 있는지 확인
    title = driver.find_element(By.CSS_SELECTOR, "#title > h1 > yt-formatted-string").text
    if title.startswith('[Playlist]'):
        # 설명란에서 모든 링크 추출
        links = driver.find_elements(By.CSS_SELECTOR, "#description a")
        for link in links:
            href = link.text
            # 정규 표현식을 사용하여 특정 패턴 링크만 리스트에 추가
            if re.match(r"http://bugs.kr/!([^\spd]+)$", href):
                all_links.append(href)  # 링크를 리스트에 추가
                img_list.append(f"https://img.youtube.com/vi/{video_url[32:]}/maxresdefault.jpg") # 이미지 링크 추가
                head_list.append(title)
# 결과 출력
print(len(all_links), len(img_list), len(head_list))
# 웹 드라이버 종료
driver.quit()
```

```python
import pandas as pd
dic = {'이미지' : img_list, '제목' : head_list, '벅스 링크' : all_links}
data = pd.DataFrame(dic)
data['제목'] = data['제목'].str.replace('[playlist] ', '', regex=False)
data.to_csv("beomsu_data.csv", index=False, encoding="utf-8-sig")
data
```
