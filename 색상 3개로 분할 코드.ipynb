```python
import pandas as pd

df = pd.read_csv('curation_data/큐레이션_멜론_필터링_통합_이미지링크제거.csv')

# Reset index
df.reset_index(drop=True, inplace=True)

# Create empty columns for new color and percentage values
df['curColor2'] = None
df['curPer2'] = None
df['curColor3'] = None
df['curPer3'] = None

# Loop through the unique combinations of curHead, curTitle, and curArtist
for index, row in df.iterrows():
    if index % 3 == 0:  # Only handle the first entry of each triplet
        # Fill the new columns with the appropriate values
        df.at[index, 'curColor2'] = df.at[index + 1, 'curColor']
        df.at[index, 'curPer2'] = df.at[index + 1, 'curPer']
        df.at[index, 'curColor3'] = df.at[index + 2, 'curColor']
        df.at[index, 'curPer3'] = df.at[index + 2, 'curPer']

# Drop the second and third rows of each triplet
filtered_df = df[df.index % 3 == 0].reset_index(drop=True)

# Reorder the columns to place curColor2, curPer2, curColor3, curPer3 next to curColor, curPer
columns_order = ['curColor', 'curPer', 'curColor2', 'curPer2', 'curColor3', 'curPer3', 
                 'curObject', 'curHead', 'curTitle', 'curExplain', 'curArtist']

# Reorder the dataframe based on the new columns order
reordered_df = filtered_df[columns_order]

# Display the reorganized dataframe
reordered_df
```
