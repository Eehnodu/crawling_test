```python
import pandas as pd
from scipy.spatial import distance

# 15개의 대표 색상과 그 RGB 값 정의
colors = {
    "Red": (255, 0, 0),
    "Green": (0, 128, 0),
    "Blue": (0, 0, 255),
    "Yellow": (255, 255, 0),
    "Orange": (255, 165, 0),
    "Purple": (128, 0, 128),
    "Pink": (255, 192, 203),
    "Cyan": (0, 255, 255),
    "Grey": (128, 128, 128),
    "Brown": (165, 42, 42),
    "Sky Blue": (135, 206, 235),
    "Lime": (0, 255, 0),
    "Chartreuse": (127, 255, 0),
    "Navy": (0, 0, 128),
    "Black": (0, 0, 0)
}

# 주어진 RGB 값에 가장 가까운 색상을 찾는 함수
def find_nearest_color(rgb):
    nearest_color = None
    min_distance = float('inf')
    for color, color_rgb in colors.items():
        dist = distance.euclidean(rgb, color_rgb)
        if dist < min_distance:
            min_distance = dist
            nearest_color = color
    return nearest_color

# 문자열을 파싱하여 RGB 값으로 변환하는 함수
def parse_rgb_string(rgb_string):
    rgb_string = rgb_string.strip('[]')
    return tuple(map(int, rgb_string.split(',')))

# 색감비율 컬럼의 값을 소수점 두자리까지
color['색감비율'] = color['색감비율'].apply(lambda x: round(x, 2))

# '이미지색감' 컬럼의 값을 RGB 값으로 변환
color['이미지색감'] = color['이미지색감'].apply(parse_rgb_string)

# RGB 값을 색상명으로 변환
color['이미지색감'] = color['이미지색감'].apply(find_nearest_color)


```

```python
import pandas as pd
from scipy.spatial import distance

# 15개의 대표 색상과 그 RGB 값 정의
colors = {
    "Red": (255, 0, 0),
    "Green": (0, 128, 0),
    "Blue": (0, 0, 255),
    "Yellow": (255, 255, 0),
    "Orange": (255, 165, 0),
    "Purple": (128, 0, 128),
    "Pink": (255, 192, 203),
    "Cyan": (0, 255, 255),
    "Grey": (128, 128, 128),
    "Brown": (165, 42, 42),
    "Sky Blue": (135, 206, 235),
    "Lime": (0, 255, 0),
    "Chartreuse": (127, 255, 0),
    "Navy": (0, 0, 128),
    "Black": (0, 0, 0)
}

# 주어진 RGB 값에 가장 가까운 색상을 찾는 함수
def find_nearest_color(rgb):
    nearest_color = None
    min_distance = float('inf')
    for color, color_rgb in colors.items():
        dist = distance.euclidean(rgb, color_rgb)
        if dist < min_distance:
            min_distance = dist
            nearest_color = color
    return nearest_color

# 문자열을 파싱하여 RGB 값으로 변환하는 함수
def parse_rgb_string(rgb_string):
    rgb_string = rgb_string.strip('()')
    return tuple(map(int, map(float, rgb_string.split(','))))
    
# 색감비율 컬럼의 값을 소수점 두자리까지
color_df['curPer'] = color_df['curPer'].apply(lambda x: round(x, 2))

# '이미지색감' 컬럼의 값을 RGB 값으로 변환
color_df['curColor'] = color_df['curColor'].apply(parse_rgb_string)

# RGB 값을 색상명으로 변환
color_df['색상'] = color_df['curColor'].apply(find_nearest_color)

```
