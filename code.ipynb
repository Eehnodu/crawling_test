```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 웹 드라이버 시작
driver = wb.Chrome()

# YouTube 채널 페이지 요청
driver.get("https://www.youtube.com/@essentialme/videos")
time.sleep(2)

# 스크롤 다운으로 모든 동영상 로드
body = driver.find_element(By.TAG_NAME, "body")
last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # 페이지 끝까지 스크롤 다운
    body.send_keys(Keys.END)
    time.sleep(2)  # 페이지가 로드될 때까지 충분한 시간 기다리기
    
    # 새로 로드된 페이지 높이를 가져와 이전 높이와 비교
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 동영상 썸네일 요소 찾기, YouTube의 구조가 변경되었을 수 있으니 확인 필요
thumbnails = driver.find_elements(By.CSS_SELECTOR, "#thumbnail > yt-image > img")

for index in range(len(thumbnails)):
    if index >= len(thumbnails):
        break
    
    if thumbnails[index]:
        try:
            # 썸네일이 클릭 가능할 때까지 기다림
            WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "#thumbnail > yt-image > img"))
            )
            thumbnails[index].click()
            time.sleep(3)
            
            # '더보기' 버튼 클릭
            try:
                thebogi = driver.find_element(By.CSS_SELECTOR, "#expand")
                thebogi.click()
                time.sleep(2)
            except Exception as e:
                print("더보기 버튼을 찾을 수 없습니다:", e)
            
            # 설명란에서 모든 링크 추출
            links = driver.find_elements(By.CSS_SELECTOR, "#description-inline-expander")
            for link in links:
                href = link.get_attribute("href")
                print(f"동영상 {index + 1} 링크: {href}")
            
            # 페이지 뒤로 가기
            driver.back()
            time.sleep(2)
            
            # 페이지가 다시 로드된 후 썸네일 요소를 다시 찾기
            thumbnails = driver.find_elements(By.CSS_SELECTOR, "#thumbnail > yt-image > img")
        except Exception as e:
            print(f"썸네일 클릭 중 오류 발생: {e}")

# 웹 드라이버 종료
driver.quit()
```
